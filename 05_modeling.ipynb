```python
import pandas as pd
import numpy as np
import pickle
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.grid_search import GridSearchCV
```

```python
import os
```

```python
from sklearn.metrics import roc_auc_score,accuracy_score
```

```python
os.chdir('../output_data')
```

```python
df = pd.read_csv('train_v04.csv',header=0)
```

```python
df_lr = pd.read_csv('train_lr.csv',header=0)
```

```python
df.shape
```

```python
df_lr.shape
```

# Logistic Regression Inferential Model

```python
x = df_lr.drop(['lift_ind'],axis=1)
```

```python
tuned_parameters = {'C':np.logspace(-2,3,6),'solver':['lbfgs','liblinear']}
model = GridSearchCV(LogisticRegression(),tuned_parameters,scoring='roc_auc',
                    cv=5,verbose=60)
```

```python
y = np.array(df_lr['lift_ind'])
```

```python
model.fit(x,y)
```

```python
final_model = model.best_estimator_
```

```python
final_model.fit(x,y)
```

```python
final_model = LogisticRegression(C=100.0, class_weight=None, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=100, multi_class='ovr', n_jobs=1,
          penalty='l2', random_state=None, solver='liblinear', tol=0.0001,
          verbose=0, warm_start=False)
```

```python
final_model.fit(x,y)
```

```python
y_prob = final_model.predict_proba(x)
```

```python
gini_train = roc_auc_score(y,y_prob[:,1])
```

```python
gini_train
```

```python
y_pred = model.predict(x)
accuracy_score(y,y_pred)
```

```python
os.chdir('../Models/')
```

```python
pickle.dump(final_model,open('LR.pickle','wb'))
```

```python
lr_model_columns = df_lr.drop(['lift_ind'],axis=1).columns
```

```python
os.chdir('../pickles')
```

```python
pickle.dump(lr_model_columns,open('lr_model_columns.pickle','wb'))
```

# Random Forest Inferential Model

```python
os.chdir('../Models')
```

```python
tuned_parameters={'max_features':[None,'sqrt','log2'],'min_samples_leaf':[i*.01 for i in range(1,11)]}
model = GridSearchCV(RandomForestClassifier(n_jobs=-1,n_estimators=1000,oob_score=True,random_state=42),tuned_parameters,cv=5,scoring='roc_auc',verbose=1000)
```

```python
x = df_lr.drop(['lift_ind'],axis=1)
```

```python
y = np.array(df_lr['lift_ind'])
```

```python
model.fit(x,y)
```

```python
model.best_estimator_
```

```python
model.best_score_
```

```python
final_model = model.best_estimator_
```

```python
final_model.fit(x,y)
```

```python
y_prob = final_model.predict_proba(x)
```

```python
roc_auc_score(y,y_prob[:,1])
```

```python
y_pred = final_model.predict(x)
accuracy_score(y,y_pred)
```

```python
pickle.dump(final_model,open('RF.pickle','wb'))
```
