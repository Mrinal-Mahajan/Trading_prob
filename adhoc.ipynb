```python
import pandas as pd
import numpy as np
import pickle
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.grid_search import GridSearchCV
```

```python
import os
```

```python
from sklearn.metrics import roc_auc_score,accuracy_score
```

```python
os.chdir('../output_data')
```

```python
df_lr = pd.read_csv('train_lr.csv',header=0)
```

```python
x = df_lr.drop(['lift_ind'],axis=1)
```

```python
y = np.array(df_lr['lift_ind'])
```

```python
os.chdir('../Models')
```

```python
tuned_parameters={'n_estimators':[100*i for i in range(1,11)]}
model = GridSearchCV(GradientBoostingClassifier(learning_rate=.1,min_samples_leaf=.01,max_depth=10,subsample=.8,
                                               random_state=42),tuned_parameters,cv=5,scoring='roc_auc',verbose=100)
```

```python
model.fit(x,y)
```

```python
model.best_estimator_
```

```python
model.best_score_
```
